## Make it a network

```python
def add_mentions(relevant, username):
    for row in tweets.loc[tweets['username'] == username]['mentions'].values.tolist():
        row = row.replace("['", '').replace("']", '').replace("[", '').replace("]", '').split("', '")
        if row[0] != '':
            relevant.extend(row)
    return relevant

def add_following(relevant, username):
    following = pd.read_csv('data/{}.csv'.format(username))
    following = following['username'].values.tolist()
    for user in following:
        if user in controversy_users:
            relevant.append(user)
    return relevant

def remove_duplicates(relevant):
    list_set = set(relevant) 
    relevant = (list(list_set)) 
    return relevant
```

```python
user_network = pd.DataFrame(columns=['username', 'contacts_list'])
controversy_users = tweets['username'].unique()
for username in controversy_users:
    relevant = []
    if os.path.exists('data/{}.csv'.format(username)):
        relevant = add_following(relevant, username)
    relevant = add_mentions(relevant, username)
    relevant = remove_duplicates(relevant)
    user_network = user_network.append({'username': username, 'contacts_list': relevant}, ignore_index=True)
```

## Get info about users

```python
def get_user_info(twitter_user):
    """
    An example of using the query_user_info method
    :param twitter_user: the twitter user to capture user data
    :return: twitter_user_data: returns a dictionary of twitter user data
    """
    user_info = query_user_info(user= twitter_user)
    twitter_user_data = {}
    if user_info is not None:
        twitter_user_data["user"] = user_info.user
        twitter_user_data["fullname"] = user_info.full_name
        twitter_user_data["location"] = user_info.location
        twitter_user_data["blog"] = user_info.blog
        twitter_user_data["date_joined"] = user_info.date_joined
        twitter_user_data["id"] = user_info.id
        twitter_user_data["num_tweets"] = user_info.tweets
        twitter_user_data["following"] = user_info.following
        twitter_user_data["followers"] = user_info.followers
        twitter_user_data["likes"] = user_info.likes
        twitter_user_data["lists"] = user_info.lists
    
    return twitter_user_data

twitter_user_info = []
users = tweets['username'].unique()
# for user in users:
#     info = get_user_info(user)
#     twitter_user_info.append(info)
```

```python
cols=['user', 'id','fullname','date_joined','location','blog', 'num_tweets','following','followers','likes','lists']
users_info = pd.DataFrame(twitter_user_info, index=users, columns=cols)
# users_info.index.name = "Users"
users_info.sort_values(by="num_tweets", ascending=False, inplace=True, kind='quicksort', na_position='last')

users_info.head()
```
