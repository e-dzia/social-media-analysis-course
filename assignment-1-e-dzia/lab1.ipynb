```python
import pandas as pd
import folium
from folium.plugins import HeatMap
```

```python
xl_file = pd.ExcelFile("../data/dashboard.xlsx")
stream = xl_file.parse("Stream")

xl_file = pd.ExcelFile("../data/content.xlsx")
top_hashtags = xl_file.parse("Top Hashtags")
```

```python
stream.head()
```

Add '#' so searching for hashtags would be easier

```python
top_hashtags['Hash'] = '#' + top_hashtags['Hash']
top_hashtags.head()
```

Get only tweets that contain popular hashtags:

```python
popular_stream = stream[stream['Tweet content'].str.contains('|'.join(top_hashtags['Hash']))]
```

```python
popular_stream.head()
```

```python
locations = popular_stream[['Latitude', 'Longitude']]
```

```python
m = folium.Map(location=[53.0, -2.0], zoom_start=6)

HeatMap(data=locations, max_zoom=3, radius=3, blur=4).add_to(m)
```
