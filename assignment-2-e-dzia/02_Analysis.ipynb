```python
import pickle

with open('tweets.pkl', 'rb') as f:
    tweets_per_user = pickle.load(f)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

## Task 2.A

A. Show the number of tweets in time (x axis) for candidates using any line plot; show the number of re-tweets in time (x axis) for candidates using any line plot

```python
users = ['ProtasiewiczJ', 'RozeckaPL', 'K_Smiszek', 'TudujKrzysztof', 'SchetynadlaPO']
```

```python
fig, ax = plt.subplots(figsize=(15, 5))
for user, tweets in zip(users, tweets_per_user):
    tweets = tweets[tweets.username == user]
    grp = tweets.groupby(pd.Grouper(key='timestamp', freq='W'))['tweet_id'].count()
    ax = grp.plot(ax=ax, kind='line', label=user,
                      style='-', title='Number of original tweets in time')
    plt.xlabel('Time')
    plt.ylabel('Number of tweets')
    plt.legend()
```

```python
fig, ax = plt.subplots(figsize=(15, 5))
for user, tweets in zip(users, tweets_per_user):
    tweets = tweets[tweets.username != user]
    grp = tweets.groupby(pd.Grouper(key='timestamp', freq='W'))['tweet_id'].count()
    ax = grp.plot(ax=ax, kind='line', label=user,
                      style='-', title='Number of retweets in time')
    plt.xlabel('Time')
    plt.ylabel('Number of retweets')
    plt.legend()
```

```python
fig, ax = plt.subplots(figsize=(15,5))
for user, tweets in zip(users, tweets_per_user):
    tweets = tweets[(tweets.username == user) & (tweets.timestamp >= datetime.datetime(2019, 1, 1))]  
    grp = tweets.groupby(pd.Grouper(key='timestamp', freq='W'))['tweet_id'].count()
    ax = grp.plot(ax=ax, kind='line', label=user,
                      style='-', title='Number of original tweets in time in 2019')
    plt.xlabel('Time')
    plt.ylabel('Number of tweets')
    plt.legend()
```

## Task 2.B

B. Show the popularity of tweets in time (number of likes, number of retweets) - likes assigned to tweet's publication date

```python
fig, ax = plt.subplots(figsize=(15, 5))
for user, tweets in zip(users, tweets_per_user):
    tweets = tweets[tweets.username == user]
    grp1 = tweets.groupby(pd.Grouper(key='timestamp', freq='W'))['likes'].sum()
    grp2 = tweets.groupby(pd.Grouper(key='timestamp', freq='W'))['retweets'].sum()
    grp = grp1 + grp2
    ax = grp.plot(ax=ax, kind='line', label=user,
                      style='-', title='Sum of likes and retweets of original tweets in time')
    plt.xlabel('Time')
    plt.ylabel('Sum of likes and retweets')
    plt.legend()
```

```python
fig, ax = plt.subplots(figsize=(15, 5))
for user, tweets in zip(users, tweets_per_user):
    tweets = tweets[(tweets.username == user) & (tweets.timestamp >= datetime.datetime(2019, 1, 1))]
    grp1 = tweets.groupby(pd.Grouper(key='timestamp', freq='W'))['likes'].sum()
    grp2 = tweets.groupby(pd.Grouper(key='timestamp', freq='W'))['retweets'].sum()
    grp = grp1 + grp2
    ax = grp.plot(ax=ax, kind='line', label=user,
                      style='-', title='Sum of likes and retweets of original tweets in time in 2019')
    plt.xlabel('Time')
    plt.ylabel('Sum of likes and retweets')
    plt.legend()
```

```python
fig, ax = plt.subplots(figsize=(15, 5))
for user, tweets in zip(users, tweets_per_user):
    tweets = tweets[(tweets.username == user) & (tweets.timestamp >= datetime.datetime(2019, 10, 1))]
    grp1 = tweets.groupby(pd.Grouper(key='timestamp', freq='D'))['likes'].sum()
    grp2 = tweets.groupby(pd.Grouper(key='timestamp', freq='D'))['retweets'].sum()
    grp = grp1 + grp2
    ax = grp.plot(ax=ax, kind='line', label=user,
                      style='-', title='Sum of likes and retweets of original tweets in time in October 2019')
    plt.xlabel('Time')
    plt.ylabel('Sum of likes and retweets')
    plt.legend()
```

## Task 2.C

C. Show the distribution of activity in the scope of a day (on an hourly basis) and week (on a daily basis) for tweeting and retweeting activities (both)

```python
fig, ax = plt.subplots(figsize=(15, 5))
for user, tweets in zip(users, tweets_per_user):
    tweets = tweets[(tweets.timestamp >= datetime.datetime(2019, 10, 7)) & 
                    (tweets.timestamp <= datetime.datetime(2019, 10, 14))]
    grp = tweets.groupby(pd.Grouper(key='timestamp', freq='D'))['tweet_id'].count()
    ax = grp.plot(ax=ax, kind='line', label=user,
                      style='.-', title='Number of all tweets in time in a week preceding elections')
    plt.xlabel('Time')
    plt.ylabel('Num of tweets and retweets')
    plt.legend()
```

```python
fig, ax = plt.subplots(figsize=(15, 5))
for user, tweets in zip(users, tweets_per_user):
    tweets = tweets[(tweets.timestamp >= datetime.datetime(2019, 10, 11)) & 
                    (tweets.timestamp <= datetime.datetime(2019, 10, 12))]
    grp = tweets.groupby(pd.Grouper(key='timestamp', freq='H'))['tweet_id'].count()
    ax = grp.plot(ax=ax, kind='line', label=user,
                      style='.-', title='Number of all tweets in time on 11th Oct 2019')
    plt.xlabel('Time')
    plt.ylabel('Num of tweets and retweets')
    plt.legend()
```
