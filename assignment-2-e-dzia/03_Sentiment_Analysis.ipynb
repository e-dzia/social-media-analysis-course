```python
import pickle

with open('tweets.pkl', 'rb') as f:
    tweets_per_user = pickle.load(f)
```

```python
import pandas as pd
from polyglot.text import Text
import matplotlib.pyplot as plt
import datetime
```

```python
users = ['ProtasiewiczJ', 'RozeckaPL', 'K_Smiszek', 'TudujKrzysztof', 'SchetynadlaPO']
```

```python
def get_sentiment(blob):
    text = Text(blob)
    text.language = 'pl'
    return text.polarity
```

```python
tweets_sentiment = []
for user, tweets in zip(users, tweets_per_user):
    tweets = tweets[tweets.timestamp >= datetime.datetime(2019, 1, 1)]
    grp = tweets[['text', 'timestamp']]
    grp['text'] = grp['text'].apply(get_sentiment)
    grp = grp.groupby(pd.Grouper(key='timestamp', freq='W'))['text'].mean()
    tweets_sentiment.append(grp)

tweets_sentiment[0].tail()
```

```python
fig, ax = plt.subplots(figsize=(15, 5))
for user, tweets in zip(users, tweets_per_user):
    tweets = tweets[tweets.timestamp >= datetime.datetime(2019, 1, 1)]
    grp = tweets[['text', 'timestamp']]
    grp['text'] = grp['text'].apply(get_sentiment)
    grp = grp.groupby(pd.Grouper(key='timestamp', freq='W'))['text'].mean()
    ax = grp.plot(ax=ax, kind='line', label=user,
                      style='.-', title='Sentiment of all tweets in time in a week preceding elections')
    ax.axhline(y=0, color='k', linestyle='--')
    ax.set_xlim(datetime.datetime(2019, 1, 1), datetime.datetime(2019, 10, 16))
    ax.set_ylim(-1, 1)
    plt.xlabel('Time')
    plt.ylabel('Num of tweets and retweets')
    plt.legend()
```
